```python
import openpyxl
from datetime import datetime

#新しいワークブックを作成
workbook = openpyxl.Workbook()

#アクティブなシートを取得
sheet = workbook.active

# 現在の日付（年月日）を取得して、変数todayに代入する
today = datetime.today()

# 固定値の設定
sheet['A1'] = '請求書'
sheet['A3'] = '株式会社ABC'
sheet['A4'] = '〒101-0022 東京都千代田区神田練塀町300'
sheet['A5'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['A6'] = '担当者名:鈴木一郎 様'
sheet['E3'] = 'No.'
sheet['E4'] = '日付'
sheet['F3'] = '0001'

# 現在の日付（年月日）を取得して、変数todayに代入する
today = datetime.today()

# 変動値の設定
sheet['F4'] = today.strftime('%Y/%m/%d')

#表のヘッダー行を書き込む
header = ['商品名', '数量', '単価', '金額']
sheet.append(header)
sheet.insert_rows(7, 2)

#表のデータを書き込む
data = [
    ['商品A',2, 10000, 20000],
    ['商品B',1, 15000, 15000]
]

#繰り返し処理でrowに格納し1行ずつ追加
for row in data:
  sheet.append(row)

# 金額(売上)の合計を計算
# dataはリストなので、4列目の値を合計します
total = sum(row[3] for row in data)
sheet.append(['', '', '', total])

# 行を追加（合計）
sheet.append(['合計', '', '', total])

# 行を追加（消費税）
tax_rate = total/10
sheet.append(['消費税', '', '', tax_rate])

# 行を追加（合計+消費税）
tax_included =  total + tax_rate
sheet.append(['税込合計', '', '', tax_included])

#行の挿入
sheet.insert_rows(13, 1)

#列の挿入
sheet.insert_cols(1, 1)

# ファイル名の設定・ファイルの保存
filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"
workbook.save(filename)
```
